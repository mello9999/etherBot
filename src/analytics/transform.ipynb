### Fee data in transactions on Ethereum Blocks 

```python
transaction_data2.info()
```

```python
feeData2 = transaction_data2[['blockNumber', 'transaction', 'gasUsed', 'gasPrice']].copy()
feeData2.sample(5)
```

```python
# gasPrice [Ether] convert to gasPrice [USD]
gasPriceUSD = feeData2['gasPrice'] / 1000000000000000000
# 1 USD = 33.48 THB
gasPriceTHB = gasPriceUSD * float(33.48)
# Value of Network Fee structure > totalFee = networkFee + itx 
# Value of Network Fee structure unit is Ether
networkFee = feeData2['gasUsed'] * feeData2['gasPrice'] 
itx = feeData2['gasUsed'] * 0.00000002  # Gas Price: 0.00000002 Ether (20 Gwei)
totalFee = networkFee + itx 

# Change units from 'Ether' to 'USD' and 'THB' 
# unit is USD
# 1 ETH = 3863.66 USD
totalFee_priceUSD = totalFee / 1000000000000000000

# unit is THB
# 1 ETH = 129,355.34 THB
# 1 USD = 33.48 THB
totalFee_priceTHB = totalFee_priceUSD * float(33.48)
```

```python
feeData2['networkFee'] = networkFee
feeData2['itx_commission'] = itx
feeData2['totalFeePrice'] = totalFee
feeData2['totalFeePriceUSD'] = totalFee_priceUSD
feeData2['totalFeePriceTHB'] = totalFee_priceTHB
```
